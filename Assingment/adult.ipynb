```python
df.isnull().sum()
```

```python
workclass_null = (df["workclass"].isnull().sum() / df.shape[0] ) * 100
occupation_null = (df["occupation"].isnull().sum() / df.shape[0] ) * 100
native_country_null = (df["native-country"].isnull().sum() / df.shape[0] ) * 100
null_data = (df.isnull().sum().sum() / df.size) * 100

print(f"Workclass null: {workclass_null:.2f}% \nOccupation null: {occupation_null:.2f}% \nNative-Country null: {native_country_null:.2f}%")
print(f"Total null values: {null_data:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Data for visualization
categories = ["Export Value ($ Billion)", "Industry Growth (%)", "Employment (People)", "Multinational Companies"]
values = [5.21, 61, 32000, 88]

# Creating the visualization
fig, ax = plt.subplots(figsize=(10, 6))

# Bar chart for key metrics
ax.barh(categories, values, color=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'])

# Adding title and labels
ax.set_title("Current Situation of Costa Rica's Medical Device Sector")
ax.set_xlabel("Values")

# Annotating the bars with values
for index, value in enumerate(values):
    ax.text(value, index, str(value), ha='left', va='center', fontsize=12, color='black')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the enhanced visualization
metrics = {
    "Export Value ($ Billion)": 5.21,
    "Industry Growth (%)": 61,
    "Employment (Thousands)": 32,
    "Multinational Companies": 88,
    "Women in Workforce (%)": 52,
}

# Create a new figure
fig, ax = plt.subplots(figsize=(12, 8))

# Custom colors
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#64B5CD']

# Creating a horizontal bar plot
ax.barh(list(metrics.keys()), list(metrics.values()), color=colors, edgecolor='black')

# Title and labels
ax.set_title("Key Metrics of Costa Rica's Medical Device Sector", fontsize=16, weight='bold')
ax.set_xlabel("Values", fontsize=12)

# Annotating the bars with values
for index, value in enumerate(metrics.values()):
    ax.text(value + 1, index, str(value), ha='left', va='center', fontsize=12, color='black', weight='bold')

# Improve the layout
plt.tight_layout()

# Display the refined plot
plt.show()
```
